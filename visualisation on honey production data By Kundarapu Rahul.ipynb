#  3. Display the percentage distribution of the data in each year using the pie chart.

```python
year = df.groupby('year').size()
year.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage Distribution of Data in Each Year')
plt.show()
```

#  4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
sns.set_style('whitegrid')
sns.distplot(df['priceperlb'], kde = False, color ='red', bins = 5)
```

```python
sns.displot(df, x="priceperlb", kde=True)
plt.show()
```

```python
from scipy.stats import skew
# to calculate skewness of a specific column
column_skew = skew(df['priceperlb'])

print("Skewness :", column_skew)
```

```python
#from the above plot, we can see that the distribution of the variable "price per lb" is not normally distributed.
#data is skewed towards right.we can see in the graph it is tailing towards the right.
#A right-skewed graph indicates that there are more low values in the data set than high values.
```

#  5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='numcol', y='prodvalue', data=df)
plt.title('Scatterplot of numcol vs prodvalue')
plt.xlabel('numcol')
plt.ylabel('prodvalue')
plt.grid(True)
plt.show()
```

```python
#From the scatterplot, we can see that there is a positive linear relationship between the variables "numcol" and "prodval".
#As the number of colonies increases, the production value also tends to increase.
#However, there is some variability in the data and the relationship is not perfectly linear
```

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Boxplot of prodvalue across different years')
plt.xlabel('Year')
plt.ylabel('Prodvalue')
plt.grid(True)
plt.show()
```

```python
#in the above box plot each column values have outliers with respect to different year columns and production values.
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
A = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']
sns.pairplot(df[A], hue="year")
plt.show()
```

```python
#From the pairplot, we can see the following relationships:
#There is a positive linear relationship between "numcol" and "total prod". As the number of colonies increases, the total production also tends to increase.
#There is a positive linear relationship between "yield percol" and "total prod". As the yield per colony increases, the total production also tends to increase.
#There is a positive linear relationship between "total prod" and "prodvalue". As the total production increases, the production value also tends to increase.
#There is a positive linear relationship between "numcol" and "prodvalue". As the number of colonies increases, the production value also tends to increase.
#In addition, we can see that the relationships between these variables remain relatively consistent across different years. This suggests that these 
#relationships are relatively stable over time and may be useful for predicting future outcomes.
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue') 

```python
B = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
correlation=df[B]
corr_matrix = correlation.corr()
sns.heatmap(corr_matrix, annot= True);
```

```python
#There is a strong positive correlation (0.91) between "numcol" and "total prod".
#There is a moderate positive correlation (0.62) between "yield percol" and "total prod".
#There is a strong positive correlation (0.85) between "total prod" and "prodvalue".
#There is a moderate negative correlation (-0.50) between "stocks" and "price per lb".
#There is a moderate negative correlation (-0.61) between "price per lb" and "prodvalue".
#In addition, we can see that there are no strong negative correlations between any of the variables.
#This suggests that there are no variables that are strongly negatively related to each other.
#Overall, the heatmap provides a comprehensive visualization of the correlation values between multiple columns.
```
