```python
#Create a new column as 'published_month' using 
#the column published_at (display the months only)
youtube_data['published_month']=youtube_data['published_at'].str[5:7]
youtube_data[['published_month']]
```

```python
#Replace the numbers in the column published_month 
#as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....
month={'01':'Jan','02':'Feb','03':'Mar','04':'Apr','05':'May','06':'Jun','07':'Jul','08':'Aug','09':'Sep','10':'Oct','11':'Nov','12':'Dec'}
month
```

```python
#Find the number of videos published each month 
#and arrange the months in a decreasing order based on the video count.
youtube_data['published_month']=youtube_data['published_month'].map(month)
youtube_data['published_month']
```

```python
#Does the number of views have any effect on how many people disliked the video?
#Support your answer with a metric and a plot.
youtube_data['dislikes rate']=youtube_data['dislikes']/youtube_data['view_count']*100
```
